### <span style="color:blue">Problem statement</span>

La resolución de laberintos y encontrar la ruta óptima han sido unos de los problemas más comunes en las áreas de robótica y optimización, es por esto que con este proyecto se busca diseñar e implementar un algoritmo que pueda resolverlos de manera rápida con un robot autónomo y que posiblemente pueda llegar a participar en competiciones de tal índole.

Se tiene como objetivo la implementación y diseño de un algoritmo a un robot autónomo para la resolución de un laberinto de paredes; un sistema autónomo para el robot en un sistema embebido (Adecuado a la normas IEEE para las competencias) y un diseño propio con especificaciones adecuadas de una competencia de micromouse.
